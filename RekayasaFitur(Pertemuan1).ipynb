```python

# REKAYASA FITUR - PERTEMUAN 1
# TOPIK: Pengenalan Data dan Fitur

# Tujuan:
# 1. Mahasiswa memahami konsep data dan fitur dalam Machine Learning.
# 2. Mahasiswa mampu melakukan eksplorasi awal dataset (EDA).
# 3. Mahasiswa dapat menampilkan informasi dasar dataset.


!pip install pandas matplotlib seaborn --quiet

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Atur tampilan plot
sns.set(style="whitegrid")
plt.rcParams['figure.figsize'] = (8,4)

# ==============================================================
# ðŸ“¥ Bagian 2: Import Dataset
# ==============================================================
# Kita akan menggunakan dataset Titanic dari GitHub

url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
data = pd.read_csv(url)

# Tampilkan 5 baris pertama
data.head()
```

```python
# ==============================================================
# Bagian 3: Eksplorasi Awal Dataset
# ==============================================================

# Melihat dimensi data
print("Jumlah baris dan kolom:", data.shape)

# Melihat tipe data setiap kolom
print("\nTipe data tiap kolom:")
print(data.dtypes)

# Melihat ringkasan statistik numerik
print("\nStatistik deskriptif:")
print(data.describe())

# Mengecek missing values
print("\nJumlah nilai kosong per kolom:")
print(data.isnull().sum())

# ==============================================================
# Insight:
# - Kolom apa yang paling banyak missing value?
# - Kolom mana yang numerik dan mana yang kategorikal?
# ==============================================================
```

```python
# ==============================================================
# Bagian 3: Eksplorasi Awal Dataset
# ==============================================================

# Melihat dimensi data
print("Jumlah baris dan kolom:", data.shape)

# Melihat tipe data setiap kolom
print("\nTipe data tiap kolom:")
print(data.dtypes)

# Melihat ringkasan statistik numerik
print("\nStatistik deskriptif:")
print(data.describe())

# Mengecek missing values
print("\nJumlah nilai kosong per kolom:")
print(data.isnull().sum())

# ==============================================================
# Insight:
# - Kolom apa yang paling banyak missing value?
# - Kolom mana yang numerik dan mana yang kategorikal?
# ==============================================================
```

```python
# ==============================================================
# Bagian 4: Visualisasi
# ==============================================================

# Distribusi umur penumpang
sns.histplot(data['Age'], kde=True)
plt.title('Distribusi Umur Penumpang Titanic')
plt.show()

# Perbandingan jumlah penumpang yang selamat berdasarkan gender
sns.countplot(x='Survived', hue='Sex', data=data)
plt.title('Kelangsungan Hidup Berdasarkan Gender')
plt.show()

# ==============================================================
# Insight:
# - Apa yang dapat disimpulkan dari distribusi umur?
# - Apakah ada perbedaan signifikan antara pria dan wanita dalam keselamatan?
# ==============================================================
```
